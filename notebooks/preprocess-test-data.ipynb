```python
import os 
import torch
import numpy as np
import pandas as pd

from segment.data.preprocess_data import preprocess_volume
from segment.data.preprocess.standard_preprocess import Preprocess3D
from segment.utils.file_utils import read_yaml_file, write_json_file
from segment.utils.utils import get_progress, multiprocess
```

**Prepare Test Data**

```python
parent_path = "/content/drive/MyDrive/Seg3D/KiTS2019/kits19/data/"

imaging_path = [parent_path + f"case_00{num}/imaging.nii.gz" for num in range(210, 300)]
case_id = [f"case_00{num}" for num in range(210, 300)]
```

```python
test_df = pd.DataFrame({"case_id": case_id, "img_path": imaging_path})
test_df
```

```python
test_df.loc[0, "img_path"]
```

```python
data_dict = test_df.to_dict(orient="records")
data_dict[0]
```

```python
config_path = "/content/drive/MyDrive/Seg3D/KiTS2019/kits19/configs/preprocess_pipeline.yaml"
dataset_config_path = "/content/drive/MyDrive/Seg3D/KiTS2019/kits19/configs/segment_pipeline.yaml"

configs = read_yaml_file(config_path)["preprocess"]
dataset_configs = read_yaml_file(config_path)["create_dataset"]
```

```python
preprocess = Preprocess3D(
    data=data_dict,
    configs=configs,
    vol_path="/content/drive/MyDrive/Seg3D/KiTS2019/kits19/output/test_data/vol_v1.1",
    seg_path="/content/drive/MyDrive/Seg3D/KiTS2019/kits19/output/test_data/vol_v1.1",
)
```

```python
preprocessed_test_data =  list(map(preprocess.create_one_test_item, get_progress(data_dict)))
```

```python
case_id = [item["case_id"] for item in preprocessed_test_data]
cropped_vol = [item["vol"] for item in preprocessed_test_data]

vol_path="/content/drive/MyDrive/Seg3D/KiTS2019/kits19/output/test_data/vol_v1.1"
cropped_vol_path = [os.path.join(vol_path, f"{id_}_imaging.nii.gz") for id_ in case_id]
[np.save(vol_path, vol) for vol_path, vol in zip(cropped_vol_path, cropped_vol)]
```
